## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Data Wrangling - Cleaning & Preparing Dataset for Analysis

# 1. Dropping unnecessary columns (IDs, agent names) as they don’t contribute to CSAT analysis.
df_cleaned = df.drop(columns=['Unique id', 'Order_id', 'Agent_name', 'Supervisor', 'Manager'])

# 2. Handling Missing Values
# - Filling missing numerical values with median (better for outliers)
df_cleaned['Item_price'].fillna(df_cleaned['Item_price'].median(), inplace=True)
df_cleaned['connected_handling_time'].fillna(df_cleaned['connected_handling_time'].median(), inplace=True)

# - Filling categorical missing values with 'Unknown' where relevant
for col in ['Customer_City', 'Product_category', 'Customer Remarks', 'order_date_time']:
    df_cleaned[col].fillna('Unknown', inplace=True)

# 3. Converting Date Columns to Datetime Format
date_cols = ['order_date_time', 'Issue_reported at', 'issue_responded', 'Survey_response_Date']
for col in date_cols:
    df_cleaned[col] = pd.to_datetime(df_cleaned[col], errors='coerce')

# 4. Creating New Features for Analysis
# - Extracting Hour, Day, and Month from `Issue_reported at`
df_cleaned['Issue_reported_Hour'] = df_cleaned['Issue_reported at'].dt.hour
df_cleaned['Issue_reported_Day'] = df_cleaned['Issue_reported at'].dt.day
df_cleaned['Issue_reported_Month'] = df_cleaned['Issue_reported at'].dt.month

# 5. Encoding Categorical Variables (for better visual grouping)
df_cleaned['Agent Shift'] = df_cleaned['Agent Shift'].astype('category').cat.codes
df_cleaned['Tenure Bucket'] = df_cleaned['Tenure Bucket'].astype('category').cat.codes
df_cleaned['category'] = df_cleaned['category'].astype('category').cat.codes

# 6. Removing Outliers in Item Price (beyond 99th percentile)
price_threshold = df_cleaned['Item_price'].quantile(0.99)
df_cleaned = df_cleaned[df_cleaned['Item_price'] <= price_threshold]

# 7. Ensuring dataset is ready for visualization
df_cleaned.reset_index(drop=True, inplace=True)

print(df_cleaned.head())
````

### What all manipulations have you done and insights you found?

1. Dropped unnecessary columns (Unique id, Order_id, Agent_name, Supervisor, Manager) that do not contribute to CSAT analysis.
2. Handled missing values:
  * Filled missing numerical values (Item_price, connected_handling_time) with the median to reduce the effect of outliers.
  * Replaced missing categorical values (Customer_City, Product_category, Customer Remarks, order_date_time) with "Unknown" to retain all records.
3. Converted date columns (order_date_time, Issue_reported at, issue_responded, Survey_response_Date) into datetime format for time-based trend analysis.
4. Created new time-based features (Issue_reported_Hour, Issue_reported_Day, Issue_reported_Month) to analyze trends by time.
5. Encoded categorical variables (Agent Shift, Tenure Bucket, category) into numeric codes for better visualization.
6. Removed outliers in Item_price beyond the 99th percentile to eliminate extreme values.
